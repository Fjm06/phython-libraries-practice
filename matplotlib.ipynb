```python
#colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='blue')
plt.plot(batsman['index'],batsman['RG Sharma'],color='pink')

plt.title("R sharma vs V kohli")

plt.xlabel('season')
plt.ylabel('Runs scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#410056') #using hexcodes
plt.plot(batsman['index'],batsman['RG Sharma'],color='#6f00ff')

plt.title("R sharma vs V kohli")

plt.xlabel('season')
plt.ylabel('Runs scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#410056',linestyle='solid',linewidth='3') #using hexcodes
plt.plot(batsman['index'],batsman['RG Sharma'],color='#6f00ff',linestyle='dashdot',linewidth='1') #style can be dashed or dotted aswell

plt.title("R sharma vs V kohli")

plt.xlabel('season')
plt.ylabel('Runs scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#410056',linestyle='solid',linewidth='3',marker="D",markersize='10') #using hexcodes
plt.plot(batsman['index'],batsman['RG Sharma'],color='#6f00ff',linestyle='dashdot',linewidth='1',marker="+") #style can be dashed or dotted aswell

plt.title("R sharma vs V kohli")

plt.xlabel('season')
plt.ylabel('Runs scored')
```

```python
#legend > location
plt.plot(batsman['index'],batsman['V Kohli'],color='#410056',linestyle='solid',linewidth='3',marker="D",markersize='10',label="virat") #using hexcodes
plt.plot(batsman['index'],batsman['RG Sharma'],color='#6f00ff',linestyle='dashdot',linewidth='1',marker="+",label="Sharma") #style can be dashed or dotted aswell

plt.title("R sharma vs V kohli")

plt.xlabel('season')
plt.ylabel('Runs scored')

plt.legend(loc='best')
```

```python
#limitig axes when an outlier lies in the data
price=[48000,54000,57000,49000,47000,45000,4500000]
year=[2015,2016,2017,2018,2019,2020,2021]
plt.plot(year,price)
plt.ylim(0,75000)
plt.xlim(2017,2019)
```

# BAR CHART

```python
#simple bar chart
children=[10,20,40,35]
color=['red','blue','green','yellow']
plt.barh(color,children,color="black")  # can control colors and labels
```

```python
df=pd.read_csv("/batsman_season_record.csv")
df
```

```python
plt.bar(df['batsman'],df['2015'],width=0.5)
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2,color='red') #numerical variables needed to provide a shift operation
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2)

plt.xticks(np.arange(df.shape[0]),df['batsman'])

plt.show()
```

```python
# a problem
children=[10,20,40,35]
color=['red red red red red red','blue blue blue blue','green green green green','yellow']
plt.bar(color,children,color="black")

#solution
plt.xticks(rotation='vertical')
```

```python
#stacked bar chart
plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=df['2016']+df['2015'],label='2017')

plt.legend()
plt.show()

#total runs scored by batsman in three years
```
